```python
urls = ['https://en.wikipedia.org/wiki/Bharatiya_Janata_Party',
        'https://en.wikipedia.org/wiki/All_India_Anna_Dravida_Munnetra_Kazhagam',
        'https://en.wikipedia.org/wiki/Indian_National_Congress',
        'https://en.wikipedia.org/wiki/All_India_Trinamool_Congress',
        'https://en.wikipedia.org/wiki/Biju_Janata_Dal',
        'https://en.wikipedia.org/wiki/Shiv_Sena',
        'https://en.wikipedia.org/wiki/Telugu_Desam_Party',
        'https://en.wikipedia.org/wiki/Telangana_Rashtra_Samithi',
        'https://en.wikipedia.org/wiki/Communist_Party_of_India_(Marxist)']
```

```python
x = 0
dflist = []
for url in urls:
    if x==4 or x == 7:
        data = pd.read_html(urls[x])[1]
    else:
        data = pd.read_html(urls[x])[0]
    data = data.set_index(0)
    data = data.rename_axis('Properties')
    data = data.rename(columns={1:df['Party'][x]})
    data = data.dropna()
    data[df['Party'][x]]= data[df['Party'][x]].str.split('[').str[0]
    dflist.append(data)
    x += 1
```

```python
joineddf = dflist[0].join(dflist[1], on='Properties', how='outer')
for x in range(2,len(dflist)):
    joineddf = joineddf.join(dflist[x], on='Properties', how='outer')
```
