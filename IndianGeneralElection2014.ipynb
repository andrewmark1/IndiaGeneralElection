```python
from pytrends.request import TrendReq # Pseudo API for Google Trends (https://pypi.org/project/pytrends/)
```

```python
partysummary = os.path.join('output','PartySummary.csv')
df = pd.read_csv(partysummary)
df.head(10)
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
queries_list = ['BJP', 'Congress', 'AIADMK', 'trinamool congress', 'BJD']
queries_list
pytrends.build_payload(queries_list, timeframe='today 3-m', geo='IN', gprop='news')
Party_group1 = pytrends.interest_over_time()
Party_group1.reset_index(level=0, inplace=True)
Party_group1
```
