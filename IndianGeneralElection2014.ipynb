```python
from pytrends.request import TrendReq # Pseudo API for Google Trends (https://pypi.org/project/pytrends/)
```

```python
partysummary = os.path.join('output','PartySummary.csv')
df = pd.read_csv(partysummary)
df.head(9)
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
queries_list = ['BJP', 'Congress', 'AIADMK', 'trinamool congress', 'BJD']
queries_list
pytrends.build_payload(queries_list, timeframe='today 3-m', geo='IN', gprop='news')
Party_group1 = pytrends.interest_over_time()
Party_group1.reset_index(level=0, inplace=True)
Party_group1.head()
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
queries_list = ['shiv sena', 'Telugu desam', 'TRS', 'communist']
queries_list
pytrends.build_payload(queries_list, timeframe='today 3-m', geo='IN', gprop='news')
Party_group2 = pytrends.interest_over_time()
Party_group2.reset_index(level=0, inplace=True)

Party_group2.head()
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
queries_list = ['Narendra Modi', 'Rahul Gandhi', 'MK Stalin', 'Mamata banerjee', 'Navin Patnaik']
queries_list
pytrends.build_payload(queries_list, timeframe='today 3-m', geo='IN', gprop='news')
Party_Leader1 = pytrends.interest_over_time()
Party_Leader1.reset_index(level=0, inplace=True)

Party_Leader1.head()
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
queries_list = ['Uddhav Thackeray', 'N. Chandrababu Naidu', 'K. Chandrashekar Rao', 'Sitaram Yechury']
queries_list
pytrends.build_payload(queries_list, timeframe='today 3-m', geo='IN', gprop='news')
Party_Leader2 = pytrends.interest_over_time()
Party_Leader2.reset_index(level=0, inplace=True)
Party_Leader2.head()
```

```python
# Merge Party_Leader1 and Party_Leader2
Party_Leader = pd.merge(Party_Leader1, Party_Leader2, how = 'outer' )
Party_Leader = Party_Leader.drop(['isPartial'], axis = 1)
Party_Leader = Party_Leader.set_index('date')
```

```python
Party_Leader = Party_Leader.transpose()
code = ['BJP', 'INC', 'AIADMK', 'AITC', 'BJD', 'SHS', 'TDP', 'TRS', 'CPI(M)']
Party_Leader['Code'] = code
Party_Leader = Party_Leader.set_index('Code')
Party_Leader.to_csv('Party_Leader_Ratings_googleTrends.csv')
Party_Leader
```

```python
# Merge Party_group1 and Party_group2
Party_Group = pd.merge(Party_group1, Party_group2, how = 'outer' )
Party_Group = Party_Group.drop(['isPartial'], axis = 1)

Party_Group = Party_Group.set_index('date')
Party_Group = Party_Group.transpose()
code = ['BJP', 'INC', 'AIADMK', 'AITC', 'BJD', 'SHS', 'TDP', 'TRS', 'CPI(M)']
Party_Group['Code'] = code
Party_Group = Party_Group.set_index('Code')
Party_Group.to_csv('Party_Rating_googleTrends.csv')
Party_Group
```
